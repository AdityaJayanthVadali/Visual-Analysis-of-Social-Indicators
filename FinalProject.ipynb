```python
continent_data = country.groupby('continent')[[
    'economist_democracy_index', 'populism_index', 'gini_index',
    'trust_in_science_index', 'hdi'
]].mean().reset_index()

# Set up subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle("Trust in Science vs Various Factors", fontsize=16)

# Function to add interactive hover annotations
def add_hover_annotations(ax, x_data, y_data, labels):
    annot = ax.annotate(
        "", xy=(0, 0), xytext=(15, 15),
        textcoords="offset points", ha="center",
        bbox=dict(boxstyle="round", fc="w"),
        arrowprops=dict(arrowstyle="->")
    )
    annot.set_visible(False)

    def update_annot(ind):
        pos = sc.get_offsets()[ind["ind"][0]]
        annot.xy = pos
        label_text = f"{labels[ind['ind'][0]]}: ({pos[0]:.2f}, {pos[1]:.2f})"
        annot.set_text(label_text)
        annot.get_bbox_patch().set_alpha(0.8)

    def hover(event):
        vis = annot.get_visible()
        if event.inaxes == ax:
            cont, ind = sc.contains(event)
            if cont:
                update_annot(ind)
                annot.set_visible(True)
                fig.canvas.draw_idle()
            else:
                if vis:
                    annot.set_visible(False)
                    fig.canvas.draw_idle()

    fig.canvas.mpl_connect("motion_notify_event", hover)

# Plot 1: Trust in Science vs Gini Index
sc = axes[0].scatter(continent_data["gini_index"], continent_data["trust_in_science_index"], c="blue", s=100)
axes[0].set_title("Trust in Science vs Gini Index")
axes[0].set_xlabel("Gini Index (Inequality)")
axes[0].set_ylabel("Trust in Science Index")
add_hover_annotations(axes[0], continent_data["gini_index"], continent_data["trust_in_science_index"], continent_data["continent"])

# Plot 2: Trust in Science vs Populism Index
sc = axes[1].scatter(continent_data["populism_index"], continent_data["trust_in_science_index"], c="red", s=100)
axes[1].set_title("Trust in Science vs Populism Index")
axes[1].set_xlabel("Populism Index")
axes[1].set_ylabel("Trust in Science Index")
add_hover_annotations(axes[1], continent_data["populism_index"], continent_data["trust_in_science_index"], country["continent"])

# Plot 3: Trust in Science vs Democracy Index
sc = axes[2].scatter(continent_data["economist_democracy_index"], continent_data["trust_in_science_index"], c="green", s=100)
axes[2].set_title("Trust in Science vs Democracy Index")
axes[2].set_xlabel("Democracy Index")
axes[2].set_ylabel("Trust in Science Index")
add_hover_annotations(axes[2], continent_data["economist_democracy_index"], continent_data["trust_in_science_index"], continent_data["continent"])

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.show()
```
